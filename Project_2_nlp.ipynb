```python
#Import library
from transformers import pipeline
```

```python
#Load Zero-Shot Pipeline
classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)
```

```python
#Run Zero-Shot Classification
text = "I want to learn deep learning and neural networks."

candidate_labels = ["education", "sports", "technology", "politics"]

result = classifier(text, candidate_labels)
result
```

```python
#Print Predictions
print("Text:", result["sequence"])
print("\nPredictions:")
for label, score in zip(result["labels"], result["scores"]):
    print(f"{label}: {score:.4f}")
```

```python
#Batch Zero-Shot (Multiple Inputs Together)
texts = [
    "This movie was terrible.",
    "The government passed a new law.",
    "The smartphone battery life is amazing.",
]

labels = ["entertainment", "technology", "politics", "sports"]

results = classifier(texts, labels, multi_label=False)
results
```
